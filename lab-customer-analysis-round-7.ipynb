```python
def create_linear_model(X,y):
    '''
    This function creates a linear model using X to predict y

    Inputs:
      X -> pd.DataFrame
      y -> pd.Series

    Outputs:
      lm -> scikit_learn linear model
      r2 -> the r2 of the model (float)
      mse -> mse of the model (float)
      
    Libraries:
        from sklearn.linear_model import LinearRegression
        from sklearn.metrics import mean_squared_error, r2_score
    '''

    lm = LinearRegression()
    lm.fit(X,y)
    print("The intercept of the model is: ",lm.intercept_)
    print("The coeficients of the model are: ",lm.coef_)
    y_pred = lm.predict(X)
    r2 = lm.score(X,y)
    mse = mean_squared_error(y_pred,y)
    print("The r2 of the model is: {:.3f}".format(r2))
    print("The MSE of the model is: {:.2f}".format(mse))
    c = list(zip(list(X.columns),list(lm.coef_))) 
    print(c)   

    return lm, r2, mse
```

```python
def log_transfom_clean(x):
    """
    This data transformation is used to make highly skewed distributions less skewed, it tries to conform to a normal 
    distribution. 
    
    Library:
        import numpy as np
        
    Apply the function to a column:
    df['column_name'] = list(map(log_transfom_clean, df['column_name']))
    To replace missing values:
    df['column_name'] = df['column_name'].fillna(np.choose_a_function(df['column_name']))
    """
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN 
```

```python
def standard_scaler(X):
    '''
    Standardize features by removing the mean and scaling to unit variance.
        Library: from sklearn.preprocessing import StandardScaler
        Input: enter the independent data to be processed 
        Output: pandas dataframe standardized
    '''
    transformer = StandardScaler()
    transformer.fit(X)
    x_standardized = transformer.transform(X)
    x_norm = pd.DataFrame(x_standardized, columns = X.columns)
    return x_norm
```
